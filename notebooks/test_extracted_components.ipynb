```python
%load_ext autoreload
%autoreload 2

import sys

import matplotlib.pyplot as plt
import torch

sys.path.append("../")
import numpy as np

from lesion_gnn.datasets import DDR, Aptos
```

```python
aptos = Aptos("../data/aptos", mode="LESIONS", which_features="encoder", feature_layer=4, features_reduction="mean")
ddr = DDR("../data/DDR", mode="LESIONS", which_features="encoder", feature_layer=4, variant="train")
```

```python
classes_aptos = []
classes_ddr = []

for d in aptos:
    classes_aptos.append(int(d.y))
for d in ddr:
    classes_ddr.append(int(d.y))
```

```python
labels = ["No DR", "Moderate", "Mild", "Severe", "Prolif DR"]

unit, count = np.unique(np.asarray(classes_aptos), return_counts=True)
freq = count / count.sum()

freq = count / count.sum()
plt.bar(labels, freq)
plt.show()
```

```python
labels = ["No DR", "Moderate", "Mild", "Severe", "Prolif DR", "Ungradable"]
unit, count = np.unique(np.asarray(classes_ddr), return_counts=True)
freq = count / count.sum()
plt.bar(labels, freq)
plt.show()
```
